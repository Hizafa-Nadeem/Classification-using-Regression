```python
Y = np.divide(data["volumeto"],data["volumefrom"])
Y
```

```python
X = data["time"]
X

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
Xmean = np.mean(X_train)
Ymean = np.mean(Y_test)

print(Xmean, Ymean )
```

```python
temp1 = X_train - Xmean
temp1
```

```python
temp2 = Y_train - Ymean
temp2
```

```python
product = [x*y for x,y in zip(temp1,temp2)]
sigma = np.sum(product)
sigma
```

```python
divident = np.sum(np.square(temp1))
divident
```

```python
b1 = np.divide(sigma,divident)
b1
```

```python
b0 = Ymean - (Xmean * b1)
b0
```

Testing and Mean Square Error

```python
y_testing = b0 + (b1 * X_test)
y_testing

diff= [x-y for x,y in zip(y_testing,Y_test)]
mean_square_error= np.sum(np.square(diff))/ len(Y)
mean_square_error
```

Plot on Test Data

```python
plt.scatter(X_test, Y_test, color = "m", marker = "o",s= 30)
plt.plot(X_test,y_testing , color = "g")

plt.xlabel('time test data')
plt.ylabel('average test data')

# function to show plot
plt.show()
```

Predictions

```python
y_pred = b0 + (b1 * X)
y_pred
```

Linear Regression Plot on complete Dataset

```python
plt.scatter(X, Y, color = "m", marker = ".",s= 30)
plt.plot(X, y_pred, color = "r")

plt.xlabel('time')
plt.ylabel('average')

# function to show plot
plt.show()

```

```python
X3 = X2[['time']]


X3 = (X3 - X3.mean()) / (X3.max())

ones= [1] * len(X3)
X3.insert(0, "x0",ones , allow_duplicates=True)

X3
```

```python
Y3 = Y
Y3 = (Y3 - Y3.mean()) / (Y3.max())
Y3
```

```python
X3_train, X3_test, Y3_train, Y3_test = train_test_split(
    X3, Y3, test_size=0.2, random_state=42)

```

```python
X3_train
```

```python
X3_test
```

```python
Y3_train
```

```python
Y3_test
```

```python
theeta = rand(2)
lr = 0.3
m = len(X3_train)
iterations = 100
theeta
```

```python
def calculate_error(theeta):
    HQ = np.dot(X3_test,theeta)
    loss= np.sum(np.square( HQ - Y3_test)) / len(Y3_test)
    return loss
```

```python
error_array = []
error = calculate_error(theeta)
print("error: "+ str(error))
error_array.append(error)

for i in range(iterations):
    j = 0
    HQ = np.dot(X3_train,theeta)
    for column in X3:
        gradient =  (lr/m * ( np.sum(np.multiply((np.subtract(HQ,Y3_train)),X3_train[column]))))
        theeta[j] = theeta[j]-gradient
        j+=1
    print("theeta: "+ str(theeta))

    error = calculate_error(theeta)
    print("error: "+ str(error))
    error_array.append(error)

theeta
```

PLOT

```python
iterations_array = []
iterations_array = range(1,iterations+2)

print(error_array)

plt.plot(iterations_array,error_array, color = "g")

plt.xlabel('iterations')
plt.ylabel('loss')

# function to show plot
plt.show()
```
